```python
import cv2

def gravar_video(nome_arquivo="saida.avi", fps=15.0, largura=100, altura=75, duration=10):
    # Abre a webcam (0 = primeira câmera)
    captura = cv2.VideoCapture(1)

    # Define o codec de vídeo e cria o objeto de gravação
    quatrocc = cv2.VideoWriter_fourcc(*'XVID')  # Codec XVID
    """
    Motion JPEG	cv2.VideoWriter_fourcc(*'MJPG')	.avi	Good compatibility, large file size.
    MPEG-4	cv2.VideoWriter_fourcc(*'XVID')	.avi	Common on Windows, decent compression.
    MPEG-4 (DivX)	cv2.VideoWriter_fourcc(*'DIVX')	.avi	Similar to XVID, widely supported.
    """
    saida = cv2.VideoWriter(nome_arquivo, quatrocc, fps, (largura, altura))
    k = 1
    
    #CARREGAR MODELO TREINADO (pickle)
    
    while True:
    #while k < duration*fps:
        ret, frame = captura.read()
        if not ret:
            print("Não foi possível acessar a câmera.")
            break

        # Redimensiona para garantir o tamanho compatível com treino
        frame = cv2.resize(frame, (largura, altura))

        # Mostra o vídeo na tela original
        #cv2.imshow("Visão computacional (pressione 'q' para parar)", frame)
        
        # AQUI VIRIA O CÓDIGO PARA PREDIZER A CATEGORIA DA IMAGEM
        # NECESSÁRIO PARA TODOS OS FRAMES?
        
        #print(f'frame {k}: ' + str(type(frame)) + 'with shape: ' + str(frame.shape))
        #75 columns x 100 rows
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        #print(f'frame {k}: ' + str(type(frame_gray)) + 'with shape: ' + str(frame_gray.shape))

        #cv2.imshow("Visão computacional GRAYSCALE (pressione 'q' para parar)", frame_gray)
        
        output = classifier.predict([frame_gray.flatten()])
        print('Class: ' + str(output))

        # Grava o frame
        #saida.write(frame)
        
        k+=1

        # Pressione 'q' para sair
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Libera os recursos
    captura.release()
    saida.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    gravar_video()
```
